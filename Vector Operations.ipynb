# Vector Operation Part-1

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-3,-2, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

# Addition Of Two Vector

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 + vector_2

print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

# Subtraction of 2 vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 - vector_2

print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-1,5, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

# Multiplying a vector by a scalar

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = 2 * vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,4,6, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = -0.5 * vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-1,-1.5, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.array([2,3])

vector_2 = np.array([4,4])

cross_product = np.cross(vector_1, vector_2)
```

```python
print(cross_product)
```

# Projection Of 'A' Vector On 'B' Vector

```python
a = np.array([2, 5])

v = np.array([8, -6])

# Magnitude of 'V' vector
magnitude_v = np.sqrt(sum(v**2))

proj_a_on_v = (np.dot(a,v) / magnitude_v**2) * v
```

```python
print(f"Projection Of 'A' Vector On 'B' Vector {proj_a_on_v}")
```
